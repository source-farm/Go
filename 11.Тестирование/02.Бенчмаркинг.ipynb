# Бенчмаркинг

Бенчмаркинг - это проверка производительности программы. В Go функции для проверки производительности похожи на тестовые функции за тем исключением, что они начинаются на **Benchmark** и параметр функции имеет тип **\*testing.B**. Этот тип предоставляет большое количество таких же методов как и \*Testing.B, но ещё дополнительно несколько методов связанных с проведением бенчмарка и поле N, которое определяет сколько раз нужно выполнить измеряемую операцию.

В отличие от тестов бенчмарки не запускаются по умолчанию. Флаг -bench определяет какой бенчмарк нужно запустить. Это регулярное выражение, посредством которого отбираются бенчмарк функции из теста, которые нужно выполнить. Если передать ".", то будут запущены все бенчмарки: **go test -bench=.**. Флаг **-benchmem** позволяет дополнительно выводить стаститику использования памяти.
