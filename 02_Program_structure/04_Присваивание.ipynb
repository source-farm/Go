# Присваивание

Значение переменной можно присвоить с помощью операции присваивания:

```go
var value int
value = 1
```

 Каждый арифметический и побитовый бинарный оператор имеет соответствующий оператор присваивания:

```go
var num int
num += 23
```

Числовые значения можно также увеличивать и уменьшать с помощью постфиксных операторов **++** и **--**:

```go
import "fmt"
n := 9
n++
n--
fmt.Sprint(n)
```

В Go нет префиксных операторов **++** и **--**.

## Множественное присваивание(tuple assignment)

С помощью множественного присваивания можно присвоить присвоить значение нескольким переменным сразу. При этом сначала вычисляются все выражения, которые находятся в правой части, а затем происходит присваивание. Это бывает удобно при обмене значений двух переменных:

```go
import "fmt"
x := 1
y := 0
x, y = y, x
fmt.Sprint(x, y)
```

Если функция, возвращающая несколько значений, находится в правой части оператора присваивания, то в левой части должно быть такое же количество переменных какое возвращается этой функцией.  
Иногда функция возвращает error или bool для указания успешности выполнения вместе с основым значением. Кроме функций есть также три оператора, которые ведут себя подобным образом:  
 - **обращение к элементу словаря**: v, ok = m[key]
 - **проверка типа(type assertion)**: v, ok = x.(T)
 - **получение значения из канала**: v, ok = <-ch


## Присваиваемость(assignability)

Оператор присваивания является явной формой присваивания, но есть и такие случаи, когда присваивание происходит косвенно: передача параметров в функцию, присваивание оператором return значений соответствующим результирующим переменным, инициализация **сложных**(composite) типов с помощью литералов. Присваивание, является оно явным или неявным, всегда проходит, когда левая часть имеет такой же тип как и правая. В общем случае присваивание проходит, когда значение **присваиваемо**(assignable) типу переменной. Правила присваиваемости различны для разных типов. Например, значение nil может быть присвоено любой переменной интерфейсного или ссылочного типов.

С присваиваемостью также тесно связано понятие **сравнимости**(comparability): в любом сравнении первый операнд должен быть присваиваемым второму или наоборот.
