# Общие сведения

Многие объектно-ориентированные языки программирования имеют понятие интерфейса, но отличительной чертой Go является то, что при объявлении типа не нужно указывать все интерфейсы, которым он соответствует: достаточно чтобы тип определял необходимые какому-либо интерфейсу методы. Такой механизм проверки удовлетворения интерфейсу называется **неявным соответствием**. Можно создавать интерфейсы, которые соответствуют уже существующим типам без изменения этих типов.
