# Общая информация

Для того чтобы программа с большим количеством кода была поддерживаемой и понятной в Go код можно организовать в так называемые **пакеты**(package). В стандартную библиотеку Go входят более 100 пакетов различного функционала: сети, шифрование, web и т.д. Каждый пакет определяет уникальное пространство имён, что позволяет избежать конфликтов имён. Пакеты позволяют скрыть реализацию и также определить, что вообще доступно клиентам этого пакета.

**go tool** - это утилита командной строки для управления рабочим пространством и пакетами. На самом деле это целое семейство команд, которые начинаются на "go tool".
