# Команда go

**go** - это утилита командной строки, которая позволяет скачивать, делать запросы, форматировать, собирать, тестировать и устанавливать пакеты Go.

Команда go похожа по своему устройству на швейцарский нож. Различные функционал можно получить, передавая этой команде различные подкоманды, такие как run, build, fmt и др. Список всех подкоманда можно получить вызовом go help:

    build       compile packages and dependencies
    clean       remove object files
    doc         show documentation for package or symbol
    env         print Go environment information
    bug         start a bug report
    fix         run go tool fix on packages
    fmt         run gofmt on package sources
    generate    generate Go files by processing source
    get         download and install packages and dependencies
    install     compile and install packages and dependencies
    list        list packages
    run         compile and run Go program
    test        test packages
    tool        run specified go tool
    version     print Go version
    vet         run go tool vet on packages
    
Для того чтобы свести необходимость ручной настройки к минимуму команда go очень сильно завязана на принятых соглашениях. Например, зная название исходного файла на Go, команда go может определить пакет, в который входит этот файл, т.к. каждая папка содержит только один пакет, а путь импортирования для пакета соответствует иерархии папок в рабочем пространстве(workspace).

## Организация рабочего пространства

После скачивания и установки дистрибутива Go единственной настройкой, которую нужно выполнить вручную, является установка пути до папки так называемого **рабочего простанства**(workspace) Go. Для этого нужно установить переменную окружения GOPATH, например такой командой:

export GOPATH=/home/user/go

Рабочее пространство Go состоит из трёх папок: **bin**, **pkg** и **src**. В папке src находится исходный код. Каждый пакет находится в папке, имя которой относительно пути $GOPATH/src является путём импортирования для этого пакета. В папке pkg находятся скомпилированные пакеты, а в bin - запускаемые программы.

После установки дистрибутива Go автоматически устанавливается переменная окружения GOROOT, которая определяет корневую папку дистрибутива. Через эту переменную можно получить путь до всех пакетов, которые входят в стандартную библиотеку Go.

Чтобы увидеть текущие значения этих и других переменных окружения можно воспользоваться командой **go env**.

## Скачивание пакетов

Для скачивания пакетов используется команда **go get**.

## Сборка пакетов

Команда **go build** компилирует каждый переданный её в аргументах пакет. При этом, если пакет является библиотекой, то результат сразу удаляется, т.е. в этом случае проверяется, что в пакете нет ошибок. Если  же пакет называется main, то go build запускает компоновщик(linker) для того чтобы создать запускаемый файл в текущей папке. Запускаем файл получает имя из последнего сегмента пути импортирования пакета. Т.к. одна папка может содержать только один пакет, каждая запускаемая программа требует своей собственной папки.

По-умолчанию go build собирает переданные ей пакеты, а также все их зависимости, а затем удаляет весь скомпилированный код за исключением запускаемого файла, если таковой вообще есть. С увеличением количества пакетов необходимых для сборки программы возрастает и время сборки программы. Чтобы каждый раз не собирать неизменившиеся пакеты можно воспользоваться командой go install. Скомпилированные пакеты сохраняются в папке $GOPATH/pkg.

Также через переменные окружения можно управлять кросс-компиляцией приложений.
