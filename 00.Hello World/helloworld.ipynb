# Простейшая программа Go
**package main** закомментирован, т.к. с ним Jupyter Notebook не запускается.  
Эту программу можно сохранить в файл main.go и запустить командой **go run main.go**.  
Собрать запускаемый файл с названием **main** можно командой **go build main.go**.  

В Go код организуется в **пакеты**. Пакет состоит из одного или нескольких .go файлов, которые находятся в одной папке. Каждый .go файл начинается с обявления пакета, которому он принадлежит. В нашем примере выражение "package main" указывает на то, что этот файл принадлежит пакету "main". Пакет main является особенным в том смысле, что с помощью него указывается, что мы хотим собрать запускаемый файл, а не библиотеку. Внутри пакета main обязательно должна быть определена функция main, которая ничего не принимает на вход и ничего не возвращает. Это точка входа программы. Обявление импортируемых пакетов должно идти после обявления package.

```go
//package main
import "fmt"

func main() {
    fmt.Println("Hello, World!")
}
```

## Запускаем программу

```go
main()
```

## Просмотр документации
Для просмотра описания какого-либо элемента Go можно пользоваться утилитой **godoc**. Распространяется эта программа вместе с дистрибутивом Go.  
Чтобы посмотреть описание пакета fmt, можно запустить команду  
**godoc fmt**  

Описание функции Printf можно увидеть так  
**godoc fmt Printf**  

С помощью godoc можно также запустить локальный вебсервер, чтобы можно было смотреть документацию из браузера. Это можно сделать так  
**godoc --http :8000**
