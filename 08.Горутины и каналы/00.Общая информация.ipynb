# Общая информация

Go поддерживает так называемое **конкурентное программирование**(concurrent porgramming) - способ выражения программы в виде независимых единиц, которые могут(но необъязательно должны) работать параллельно. В Go существует два механизма для реализации конкурентного программирования:  
 - CSP(communicating sequential processes) - это когда значения передаются между независимыми единицами(горутинами), а переменные по большей части не выходят за границу единицы, в который они объявлены.
 - традиционный подход покотов с разделяемой памятью

Для реализации CSP механизма в Go существуют горутины(goroutines) и каналы и именно их мы и рассмотрим в этой главе.
